```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.metrics import MeanIoU
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
import nibabel as nib
from scipy.ndimage import zoom
import json
import pickle
from datetime import datetime

# Suppress TensorFlow warnings
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

# Set up GPU strategy
try:
    strategy = tf.distribute.MirroredStrategy()
    print(f"Number of devices: {strategy.num_replicas_in_sync}")
except:
    strategy = tf.distribute.get_strategy()
    print(f"Using default strategy: {strategy}")

# ============================================================================
# 1. DATA LOADING AND PREPROCESSING
# ============================================================================

def load_and_preprocess_data(image_dir, label_dir, target_shape=(128, 128)):
    """Load and preprocess NIfTI data with synthetic cancer labels"""
    
    def load_nifti(filepath):
        return nib.load(filepath).get_fdata()
    
    def resize_slice(slice_2d, target_shape):
        return zoom(slice_2d, [t / s for t, s in zip(target_shape, slice_2d.shape)], order=1)
    
    def generate_cancer_labels(label_slice):
        """Generate cancer and severity labels based on segmentation"""
        # Assume: 0=background, 1=prostate, 2=lesion/cancer
        lesion_pixels = np.sum(label_slice == 2)
        prostate_pixels = np.sum(label_slice == 1)
        
        if lesion_pixels == 0:
            cancer_label = 0  # No cancer
            severity_label = 0  # No severity
        else:
            cancer_label = 1  # Cancer present
            lesion_ratio = lesion_pixels / (prostate_pixels + 1e-8)
            
            # Severity based on lesion ratio
            if lesion_ratio < 0.03:  # Adjusted thresholds
                severity_label = 1  # Low
            elif lesion_ratio < 0.10:
                severity_label = 2  # Medium
            else:
                severity_label = 3  # High
        
        return cancer_label, severity_label
    
    images, labels, cancer_labels, severity_labels = [], [], [], []
    file_info = []
    
    image_files = sorted(os.listdir(image_dir))
    label_files = sorted(os.listdir(label_dir))
    
    print(f"Processing {len(image_files)} files...")
    
    for idx, (img_file, lbl_file) in enumerate(zip(image_files, label_files)):
        print(f"Processing {idx+1}/{len(image_files)}: {img_file}", end='\r')
        
        img_data = load_nifti(os.path.join(image_dir, img_file))
        lbl_data = load_nifti(os.path.join(label_dir, lbl_file))
        
        # Handle 4D images (take first channel if exists)
        if len(img_data.shape) == 4:
            img_data = img_data[:, :, :, 0]
        
        for slice_idx in range(img_data.shape[2]):
            img_slice = resize_slice(img_data[:, :, slice_idx], target_shape)
            lbl_slice = resize_slice(lbl_data[:, :, slice_idx], target_shape)
            
            # Skip empty slices
            if np.max(img_slice) > 0:
                images.append(img_slice)
                labels.append(lbl_slice)
                
                cancer_label, severity_label = generate_cancer_labels(lbl_slice)
                cancer_labels.append(cancer_label)
                severity_labels.append(severity_label)
                
                file_info.append({
                    'file': img_file,
                    'slice': slice_idx,
                    'cancer': cancer_label,
                    'severity': severity_label
                })
    
    print(f"\nLoaded {len(images)} slices")
    return (np.array(images)[..., np.newaxis], 
            np.array(labels), 
            np.array(cancer_labels), 
            np.array(severity_labels),
            file_info)

# ============================================================================
# 2. EXPLORATORY DATA ANALYSIS (EDA)
# ============================================================================

def perform_eda(images, labels, cancer_labels, severity_labels, file_info):
    """Comprehensive EDA of the dataset"""
    
    print("="*60)
    print("EXPLORATORY DATA ANALYSIS")
    print("="*60)
    
    # Create DataFrame for analysis
    df = pd.DataFrame(file_info)
    
    # Basic statistics
    print(f"\nDataset Overview:")
    print(f"Total slices: {len(images)}")
    print(f"Image shape: {images[0].shape}")
    print(f"Unique files: {df['file'].nunique()}")
    
    # Class distribution
    print(f"\nCancer Distribution:")
    cancer_dist = pd.Series(cancer_labels).value_counts().sort_index()
    print(cancer_dist)
    print(f"Cancer percentage: {(cancer_dist[1] / cancer_dist.sum() * 100):.2f}%")
    
    print(f"\nSeverity Distribution:")
    severity_dist = pd.Series(severity_labels).value_counts().sort_index()
    print(severity_dist)
    
    # Visualizations
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('Prostate Cancer Dataset EDA', fontsize=16)
    
    # 1. Cancer distribution
    axes[0, 0].pie(cancer_dist.values, labels=['No Cancer', 'Cancer'], autopct='%1.1f%%')
    axes[0, 0].set_title('Cancer Distribution')
    
    # 2. Severity distribution
    severity_labels_text = ['None', 'Low', 'Medium', 'High']
    axes[0, 1].bar(range(len(severity_dist)), severity_dist.values)
    axes[0, 1].set_xticks(range(len(severity_dist)))
    axes[0, 1].set_xticklabels([severity_labels_text[i] for i in severity_dist.index])
    axes[0, 1].set_title('Severity Distribution')
    axes[0, 1].set_ylabel('Count')
    
    # 3. Sample images
    sample_indices = np.random.choice(len(images), 3, replace=False)
    for i, idx in enumerate(sample_indices):
        axes[0, 2].imshow(images[idx].squeeze(), cmap='gray') if i == 0 else None
    axes[0, 2].set_title('Sample Image')
    axes[0, 2].axis('off')
    
    # 4. Pixel intensity distribution
    sample_pixels = images[::100].flatten()  # Sample for speed
    axes[1, 0].hist(sample_pixels, bins=50, alpha=0.7)
    axes[1, 0].set_title('Pixel Intensity Distribution')
    axes[1, 0].set_xlabel('Intensity')
    axes[1, 0].set_ylabel('Frequency')
    
    # 5. Segmentation classes
    unique_labels = np.unique(labels)
    label_counts = [np.sum(labels == label) for label in unique_labels]
    axes[1, 1].bar(unique_labels, label_counts)
    axes[1, 1].set_title('Segmentation Classes')
    axes[1, 1].set_xlabel('Class')
    axes[1, 1].set_ylabel('Pixel Count')
    
    # 6. Cancer by file
    cancer_by_file = df.groupby('file')['cancer'].max().reset_index()
    cancer_file_counts = cancer_by_file['cancer'].value_counts()
    axes[1, 2].bar(['No Cancer Files', 'Cancer Files'], cancer_file_counts.values)
    axes[1, 2].set_title('Files with/without Cancer')
    axes[1, 2].set_ylabel('Number of Files')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/eda_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Additional analysis
    print(f"\nDetailed Analysis:")
    print(f"Average lesion ratio in cancer cases: {df[df['cancer']==1].groupby('severity')['cancer'].count()}")
    
    return df

# ============================================================================
# 3. MODEL ARCHITECTURE
# ============================================================================

def create_multi_output_unet(input_shape, num_seg_classes=3):
    """Create U-Net with segmentation + classification outputs"""
    
    inputs = layers.Input(input_shape)
    
    # Encoder
    enc1 = layers.Conv2D(64, 3, activation='relu', padding='same')(inputs)
    enc1 = layers.Conv2D(64, 3, activation='relu', padding='same')(enc1)
    pool1 = layers.MaxPooling2D(pool_size=(2, 2))(enc1)
    
    enc2 = layers.Conv2D(128, 3, activation='relu', padding='same')(pool1)
    enc2 = layers.Conv2D(128, 3, activation='relu', padding='same')(enc2)
    pool2 = layers.MaxPooling2D(pool_size=(2, 2))(enc2)
    
    enc3 = layers.Conv2D(256, 3, activation='relu', padding='same')(pool2)
    enc3 = layers.Conv2D(256, 3, activation='relu', padding='same')(enc3)
    pool3 = layers.MaxPooling2D(pool_size=(2, 2))(enc3)
    
    enc4 = layers.Conv2D(512, 3, activation='relu', padding='same')(pool3)
    enc4 = layers.Conv2D(512, 3, activation='relu', padding='same')(enc4)
    pool4 = layers.MaxPooling2D(pool_size=(2, 2))(enc4)
    
    # Bridge (bottleneck)
    bridge = layers.Conv2D(1024, 3, activation='relu', padding='same')(pool4)
    bridge = layers.Conv2D(1024, 3, activation='relu', padding='same')(bridge)
    
    # ========== SEGMENTATION DECODER ==========
    up4 = layers.UpSampling2D(size=(2, 2))(bridge)
    up4 = layers.concatenate([up4, enc4])
    dec4 = layers.Conv2D(512, 3, activation='relu', padding='same')(up4)
    dec4 = layers.Conv2D(512, 3, activation='relu', padding='same')(dec4)
    
    up3 = layers.UpSampling2D(size=(2, 2))(dec4)
    up3 = layers.concatenate([up3, enc3])
    dec3 = layers.Conv2D(256, 3, activation='relu', padding='same')(up3)
    dec3 = layers.Conv2D(256, 3, activation='relu', padding='same')(dec3)
    
    up2 = layers.UpSampling2D(size=(2, 2))(dec3)
    up2 = layers.concatenate([up2, enc2])
    dec2 = layers.Conv2D(128, 3, activation='relu', padding='same')(up2)
    dec2 = layers.Conv2D(128, 3, activation='relu', padding='same')(dec2)
    
    up1 = layers.UpSampling2D(size=(2, 2))(dec2)
    up1 = layers.concatenate([up1, enc1])
    dec1 = layers.Conv2D(64, 3, activation='relu', padding='same')(up1)
    dec1 = layers.Conv2D(64, 3, activation='relu', padding='same')(dec1)
    
    # Segmentation output
    segmentation_output = layers.Conv2D(num_seg_classes, 1, activation='softmax', name='segmentation')(dec1)
    
    # ========== CLASSIFICATION BRANCH ==========
    # Use bridge features + encoder features for classification
    gap_bridge = layers.GlobalAveragePooling2D()(bridge)
    gap_enc4 = layers.GlobalAveragePooling2D()(enc4)
    
    # Combine multi-scale features
    combined_features = layers.concatenate([gap_bridge, gap_enc4])
    
    # Classification layers
    dense1 = layers.Dense(512, activation='relu')(combined_features)
    dropout1 = layers.Dropout(0.5)(dense1)
    dense2 = layers.Dense(256, activation='relu')(dropout1)
    dropout2 = layers.Dropout(0.3)(dense2)
    dense3 = layers.Dense(128, activation='relu')(dropout2)
    dropout3 = layers.Dropout(0.2)(dense3)
    
    # Output layers
    cancer_output = layers.Dense(2, activation='softmax', name='cancer_detection')(dropout3)
    severity_output = layers.Dense(4, activation='softmax', name='severity_classification')(dropout3)
    
    model = models.Model(inputs=inputs, 
                        outputs=[segmentation_output, cancer_output, severity_output])
    
    return model

# ============================================================================
# 4. CUSTOM METRICS AND LOSS FUNCTIONS
# ============================================================================

def dice_coefficient(y_true, y_pred, smooth=1):
    """Dice coefficient for segmentation"""
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + smooth)

def dice_loss(y_true, y_pred):
    """Dice loss for segmentation"""
    return 1 - dice_coefficient(y_true, y_pred)

def combined_segmentation_loss(y_true, y_pred):
    """Combined loss for segmentation"""
    return 0.5 * tf.keras.losses.categorical_crossentropy(y_true, y_pred) + 0.5 * dice_loss(y_true, y_pred)

# ============================================================================
# 5. TRAINING PIPELINE
# ============================================================================

def train_model(X_train, y_seg_train, y_cancer_train, y_severity_train,
                X_val, y_seg_val, y_cancer_val, y_severity_val):
    """Complete training pipeline"""
    
    print("="*60)
    print("TRAINING MULTI-OUTPUT U-NET MODEL")
    print("="*60)
    
    with strategy.scope():
        # Create model
        model = create_multi_output_unet(input_shape=(128, 128, 1), num_seg_classes=3)
        
        # Compile model
        model.compile(
            optimizer=Adam(learning_rate=1e-4),
            loss={
                'segmentation': combined_segmentation_loss,
                'cancer_detection': 'sparse_categorical_crossentropy',
                'severity_classification': 'sparse_categorical_crossentropy'
            },
            loss_weights={
                'segmentation': 1.0,
                'cancer_detection': 0.7,
                'severity_classification': 0.5
            },
            metrics={
                'segmentation': ['accuracy', MeanIoU(num_classes=3), dice_coefficient],
                'cancer_detection': ['accuracy'],
                'severity_classification': ['accuracy']
            }
        )
    
    # Print model summary
    model.summary()
    
    # Callbacks
    callbacks = [
        ModelCheckpoint(
            '/kaggle/working/best_model.h5',
            monitor='val_loss',
            save_best_only=True,
            save_weights_only=False,
            verbose=1
        ),
        EarlyStopping(
            monitor='val_loss',
            patience=15,
            restore_best_weights=True,
            verbose=1
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=8,
            min_lr=1e-7,
            verbose=1
        )
    ]
    
    # Calculate batch size
    batch_size = 16 * strategy.num_replicas_in_sync
    
    # Train model
    print(f"\nTraining with batch size: {batch_size}")
    history = model.fit(
        X_train,
        {
            'segmentation': y_seg_train,
            'cancer_detection': y_cancer_train,
            'severity_classification': y_severity_train
        },
        validation_data=(
            X_val,
            {
                'segmentation': y_seg_val,
                'cancer_detection': y_cancer_val,
                'severity_classification': y_severity_val
            }
        ),
        epochs=100,
        batch_size=batch_size,
        callbacks=callbacks,
        verbose=1
    )
    
    return model, history

# ============================================================================
# 6. EVALUATION AND METRICS
# ============================================================================

def evaluate_model(model, X_test, y_seg_test, y_cancer_test, y_severity_test):
    """Comprehensive model evaluation"""
    
    print("="*60)
    print("MODEL EVALUATION")
    print("="*60)
    
    # Get predictions
    predictions = model.predict(X_test, batch_size=32, verbose=1)
    seg_pred, cancer_pred, severity_pred = predictions
    
    # Convert predictions to classes
    cancer_pred_classes = np.argmax(cancer_pred, axis=1)
    severity_pred_classes = np.argmax(severity_pred, axis=1)
    seg_pred_classes = np.argmax(seg_pred, axis=-1)
    
    # Calculate metrics
    results = {}
    
    # 1. Cancer Detection Metrics
    print("\n1. CANCER DETECTION METRICS:")
    cancer_accuracy = np.mean(cancer_pred_classes == y_cancer_test)
    cancer_auc = roc_auc_score(y_cancer_test, cancer_pred[:, 1])
    
    print(f"Accuracy: {cancer_accuracy:.4f}")
    print(f"AUC-ROC: {cancer_auc:.4f}")
    
    print("\nClassification Report:")
    print(classification_report(y_cancer_test, cancer_pred_classes, 
                              target_names=['No Cancer', 'Cancer']))
    
    # 2. Severity Classification Metrics
    print("\n2. SEVERITY CLASSIFICATION METRICS:")
    severity_accuracy = np.mean(severity_pred_classes == y_severity_test)
    print(f"Accuracy: {severity_accuracy:.4f}")
    
    # Get the actual classes present in the test data
    unique_classes = np.unique(y_severity_test)
    severity_names = ['None', 'Low', 'Medium', 'High']
    present_severity_names = [severity_names[i] for i in unique_classes]
    
    print("\nClassification Report:")
    print(classification_report(y_severity_test, severity_pred_classes, 
                              target_names=present_severity_names,
                              labels=unique_classes))
    
    # 3. Segmentation Metrics
    print("\n3. SEGMENTATION METRICS:")
    seg_accuracy = np.mean(seg_pred_classes == y_seg_test)
    
    # Calculate IoU for each class
    from sklearn.metrics import jaccard_score
    seg_flat = y_seg_test.flatten()
    seg_pred_flat = seg_pred_classes.flatten()
    
    iou_scores = []
    for class_id in range(3):
        iou = jaccard_score(seg_flat == class_id, seg_pred_flat == class_id)
        iou_scores.append(iou)
    
    mean_iou = np.mean(iou_scores)
    
    print(f"Pixel Accuracy: {seg_accuracy:.4f}")
    print(f"Mean IoU: {mean_iou:.4f}")
    print(f"IoU per class: {iou_scores}")
    
    # Store results
    results = {
        'cancer_accuracy': float(cancer_accuracy),
        'cancer_auc': float(cancer_auc),
        'severity_accuracy': float(severity_accuracy),
        'segmentation_accuracy': float(seg_accuracy),
        'mean_iou': float(mean_iou),
        'iou_per_class': [float(x) for x in iou_scores],
        'timestamp': datetime.now().isoformat()
    }
    
    return results, predictions

def plot_training_history(history):
    """Plot training curves"""
    
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    fig.suptitle('Training History', fontsize=16)
    
    # Loss plots
    axes[0, 0].plot(history.history['loss'], label='Train')
    axes[0, 0].plot(history.history['val_loss'], label='Validation')
    axes[0, 0].set_title('Total Loss')
    axes[0, 0].legend()
    
    axes[0, 1].plot(history.history['cancer_detection_loss'], label='Train')
    axes[0, 1].plot(history.history['val_cancer_detection_loss'], label='Validation')
    axes[0, 1].set_title('Cancer Detection Loss')
    axes[0, 1].legend()
    
    axes[0, 2].plot(history.history['segmentation_loss'], label='Train')
    axes[0, 2].plot(history.history['val_segmentation_loss'], label='Validation')
    axes[0, 2].set_title('Segmentation Loss')
    axes[0, 2].legend()
    
    # Accuracy plots
    axes[1, 0].plot(history.history['cancer_detection_accuracy'], label='Train')
    axes[1, 0].plot(history.history['val_cancer_detection_accuracy'], label='Validation')
    axes[1, 0].set_title('Cancer Detection Accuracy')
    axes[1, 0].legend()
    
    axes[1, 1].plot(history.history['severity_classification_accuracy'], label='Train')
    axes[1, 1].plot(history.history['val_severity_classification_accuracy'], label='Validation')
    axes[1, 1].set_title('Severity Classification Accuracy')
    axes[1, 1].legend()
    
    axes[1, 2].plot(history.history['segmentation_accuracy'], label='Train')
    axes[1, 2].plot(history.history['val_segmentation_accuracy'], label='Validation')
    axes[1, 2].set_title('Segmentation Accuracy')
    axes[1, 2].legend()
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/training_history.png', dpi=300, bbox_inches='tight')
    plt.show()

def visualize_predictions(X_test, y_seg_test, y_cancer_test, y_severity_test, predictions, num_samples=6):
    """Visualize model predictions"""
    
    seg_pred, cancer_pred, severity_pred = predictions
    cancer_pred_classes = np.argmax(cancer_pred, axis=1)
    severity_pred_classes = np.argmax(severity_pred, axis=1)
    seg_pred_classes = np.argmax(seg_pred, axis=-1)
    
    fig, axes = plt.subplots(num_samples, 4, figsize=(16, 4*num_samples))
    fig.suptitle('Model Predictions', fontsize=16)
    
    severity_names = ['None', 'Low', 'Medium', 'High']
    cancer_names = ['No Cancer', 'Cancer']
    
    for i in range(num_samples):
        # Original image
        axes[i, 0].imshow(X_test[i].squeeze(), cmap='gray')
        axes[i, 0].set_title('Original Image')
        axes[i, 0].axis('off')
        
        # True segmentation
        axes[i, 1].imshow(y_seg_test[i], cmap='viridis')
        axes[i, 1].set_title('True Segmentation')
        axes[i, 1].axis('off')
        
        # Predicted segmentation
        axes[i, 2].imshow(seg_pred_classes[i], cmap='viridis')
        axes[i, 2].set_title('Predicted Segmentation')
        axes[i, 2].axis('off')
        
        # Classification results
        true_cancer = cancer_names[y_cancer_test[i]]
        pred_cancer = cancer_names[cancer_pred_classes[i]]
        true_severity = severity_names[y_severity_test[i]]
        pred_severity = severity_names[severity_pred_classes[i]]
        
        axes[i, 3].text(0.1, 0.8, f"True Cancer: {true_cancer}", fontsize=12, transform=axes[i, 3].transAxes)
        axes[i, 3].text(0.1, 0.6, f"Pred Cancer: {pred_cancer}", fontsize=12, transform=axes[i, 3].transAxes)
        axes[i, 3].text(0.1, 0.4, f"True Severity: {true_severity}", fontsize=12, transform=axes[i, 3].transAxes)
        axes[i, 3].text(0.1, 0.2, f"Pred Severity: {pred_severity}", fontsize=12, transform=axes[i, 3].transAxes)
        axes[i, 3].set_title('Classification Results')
        axes[i, 3].axis('off')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/model_predictions.png', dpi=300, bbox_inches='tight')
    plt.show()

# ============================================================================
# 7. MAIN EXECUTION PIPELINE
# ============================================================================

def main():
    """Main execution pipeline"""
    
    print("="*80)
    print("PROSTATE CANCER CLASSIFICATION AND SEGMENTATION PIPELINE")
    print("="*80)
    
    # 1. Load and preprocess data
    print("\n1. Loading and preprocessing data...")
    image_dir = '/kaggle/input/prostate-annotated-dataset-for-image-segmentation/Task05_Prostate/imagesTr'
    label_dir = '/kaggle/input/prostate-annotated-dataset-for-image-segmentation/Task05_Prostate/labelsTr'
    
    images, labels, cancer_labels, severity_labels, file_info = load_and_preprocess_data(image_dir, label_dir)
    
    # 2. EDA
    print("\n2. Performing EDA...")
    df = perform_eda(images, labels, cancer_labels, severity_labels, file_info)
    
    # 3. Prepare data for training
    print("\n3. Preparing data for training...")
    
    # Normalize images
    images = images / np.max(images)
    
    # Convert labels
    labels_categorical = tf.keras.utils.to_categorical(labels, num_classes=3)
    
    # Split data
    X_train, X_temp, y_seg_train, y_seg_temp, y_cancer_train, y_cancer_temp, y_severity_train, y_severity_temp = train_test_split(
        images, labels_categorical, cancer_labels, severity_labels, test_size=0.3, random_state=42, stratify=cancer_labels
    )
    
    X_val, X_test, y_seg_val, y_seg_test, y_cancer_val, y_cancer_test, y_severity_val, y_severity_test = train_test_split(
        X_temp, y_seg_temp, y_cancer_temp, y_severity_temp, test_size=0.5, random_state=42, stratify=y_cancer_temp
    )
    
    print(f"Training set: {len(X_train)} samples")
    print(f"Validation set: {len(X_val)} samples")
    print(f"Test set: {len(X_test)} samples")
    
    # 4. Train model
    print("\n4. Training model...")
    model, history = train_model(X_train, y_seg_train, y_cancer_train, y_severity_train,
                                X_val, y_seg_val, y_cancer_val, y_severity_val)
    
    # 5. Plot training history
    print("\n5. Plotting training history...")
    plot_training_history(history)
    
    # 6. Evaluate model
    print("\n6. Evaluating model...")
    results, predictions = evaluate_model(model, X_test, y_seg_test, y_cancer_test, y_severity_test)
    
    # 7. Visualize predictions
    print("\n7. Visualizing predictions...")
    visualize_predictions(X_test, y_seg_test, y_cancer_test, y_severity_test, predictions)
    
    # 8. Save everything
    print("\n8. Saving model and results...")
    
    # Save model
    model.save('/kaggle/working/prostate_multi_output_model.h5')
    
    # Save training history
    with open('/kaggle/working/training_history.json', 'w') as f:
        json.dump(history.history, f, default=str)
    
    # Save evaluation results
    with open('/kaggle/working/evaluation_results.json', 'w') as f:
        json.dump(results, f, indent=2)
    
    # Save data splits for reproducibility
    np.savez('/kaggle/working/data_splits.npz',
             X_test=X_test, y_seg_test=y_seg_test, 
             y_cancer_test=y_cancer_test, y_severity_test=y_severity_test)
    
    print("\n" + "="*80)
    print("PIPELINE COMPLETED SUCCESSFULLY!")
    print("="*80)
    print(f"Model saved: /kaggle/working/prostate_multi_output_model.h5")
    print(f"Results saved: /kaggle/working/evaluation_results.json")
    print(f"History saved: /kaggle/working/training_history.json")
    print(f"Final Results:")
    print(f"  - Cancer Detection Accuracy: {results['cancer_accuracy']:.4f}")
    print(f"  - Cancer Detection AUC: {results['cancer_auc']:.4f}")
    print(f"  - Severity Classification Accuracy: {results['severity_accuracy']:.4f}")

if __name__ == "__main__":
    main()
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
from tensorflow.keras.metrics import MeanIoU
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve
from sklearn.metrics import jaccard_score, accuracy_score, precision_recall_fscore_support
import nibabel as nib
from scipy.ndimage import zoom
import json
import pickle
from datetime import datetime

# Suppress TensorFlow warnings
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

# Set up GPU strategy
try:
    strategy = tf.distribute.MirroredStrategy()
    print(f"Number of devices: {strategy.num_replicas_in_sync}")
except:
    strategy = tf.distribute.get_strategy()
    print(f"Using default strategy: {strategy}")

# ============================================================================
# 1. DATA LOADING AND PREPROCESSING
# ============================================================================

def load_and_preprocess_data(image_dir, label_dir, target_shape=(128, 128)):
    """Load and preprocess NIfTI data with synthetic cancer labels"""
    
    def load_nifti(filepath):
        return nib.load(filepath).get_fdata()
    
    def resize_slice(slice_2d, target_shape):
        return zoom(slice_2d, [t / s for t, s in zip(target_shape, slice_2d.shape)], order=1)
    
    def generate_cancer_labels(label_slice):
        """Generate cancer and severity labels based on segmentation"""
        # Assume: 0=background, 1=prostate, 2=lesion/cancer
        lesion_pixels = np.sum(label_slice == 2)
        prostate_pixels = np.sum(label_slice == 1)
        
        if lesion_pixels == 0:
            cancer_label = 0  # No cancer
            severity_label = 0  # No severity
        else:
            cancer_label = 1  # Cancer present
            lesion_ratio = lesion_pixels / (prostate_pixels + 1e-8)
            
            # Severity based on lesion ratio
            if lesion_ratio < 0.03:  # Adjusted thresholds
                severity_label = 1  # Low
            elif lesion_ratio < 0.10:
                severity_label = 2  # Medium
            else:
                severity_label = 3  # High
        
        return cancer_label, severity_label
    
    images, labels, cancer_labels, severity_labels = [], [], [], []
    file_info = []
    
    image_files = sorted(os.listdir(image_dir))
    label_files = sorted(os.listdir(label_dir))
    
    print(f"Processing {len(image_files)} files...")
    
    for idx, (img_file, lbl_file) in enumerate(zip(image_files, label_files)):
        print(f"Processing {idx+1}/{len(image_files)}: {img_file}", end='\r')
        
        img_data = load_nifti(os.path.join(image_dir, img_file))
        lbl_data = load_nifti(os.path.join(label_dir, lbl_file))
        
        # Handle 4D images (take first channel if exists)
        if len(img_data.shape) == 4:
            img_data = img_data[:, :, :, 0]
        
        for slice_idx in range(img_data.shape[2]):
            img_slice = resize_slice(img_data[:, :, slice_idx], target_shape)
            lbl_slice = resize_slice(lbl_data[:, :, slice_idx], target_shape)
            
            # Skip empty slices
            if np.max(img_slice) > 0:
                images.append(img_slice)
                labels.append(lbl_slice)
                
                cancer_label, severity_label = generate_cancer_labels(lbl_slice)
                cancer_labels.append(cancer_label)
                severity_labels.append(severity_label)
                
                file_info.append({
                    'file': img_file,
                    'slice': slice_idx,
                    'cancer': cancer_label,
                    'severity': severity_label
                })
    
    print(f"\nLoaded {len(images)} slices")
    return (np.array(images)[..., np.newaxis], 
            np.array(labels), 
            np.array(cancer_labels), 
            np.array(severity_labels),
            file_info)

# ============================================================================
# 2. EXPLORATORY DATA ANALYSIS (EDA)
# ============================================================================

def perform_eda(images, labels, cancer_labels, severity_labels, file_info):
    """Comprehensive EDA of the dataset"""
    
    print("="*60)
    print("EXPLORATORY DATA ANALYSIS")
    print("="*60)
    
    # Create DataFrame for analysis
    df = pd.DataFrame(file_info)
    
    # Basic statistics
    print(f"\nDataset Overview:")
    print(f"Total slices: {len(images)}")
    print(f"Image shape: {images[0].shape}")
    print(f"Unique files: {df['file'].nunique()}")
    
    # Class distribution
    print(f"\nCancer Distribution:")
    cancer_dist = pd.Series(cancer_labels).value_counts().sort_index()
    print(cancer_dist)
    print(f"Cancer percentage: {(cancer_dist.get(1, 0) / cancer_dist.sum() * 100):.2f}%")
    
    print(f"\nSeverity Distribution:")
    severity_dist = pd.Series(severity_labels).value_counts().sort_index()
    print(severity_dist)
    
    # Visualizations
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('Prostate Cancer Dataset EDA', fontsize=16)
    
    # 1. Cancer distribution
    cancer_counts = [cancer_dist.get(0, 0), cancer_dist.get(1, 0)]
    axes[0, 0].pie(cancer_counts, labels=['No Cancer', 'Cancer'], autopct='%1.1f%%')
    axes[0, 0].set_title('Cancer Distribution')
    
    # 2. Severity distribution
    severity_labels_text = ['None', 'Low', 'Medium', 'High']
    axes[0, 1].bar(range(len(severity_dist)), severity_dist.values)
    axes[0, 1].set_xticks(range(len(severity_dist)))
    axes[0, 1].set_xticklabels([severity_labels_text[i] for i in severity_dist.index])
    axes[0, 1].set_title('Severity Distribution')
    axes[0, 1].set_ylabel('Count')
    
    # 3. Sample images
    sample_indices = np.random.choice(len(images), 1, replace=False)
    axes[0, 2].imshow(images[sample_indices[0]].squeeze(), cmap='gray')
    axes[0, 2].set_title('Sample Image')
    axes[0, 2].axis('off')
    
    # 4. Pixel intensity distribution
    sample_pixels = images[::100].flatten()  # Sample for speed
    axes[1, 0].hist(sample_pixels, bins=50, alpha=0.7)
    axes[1, 0].set_title('Pixel Intensity Distribution')
    axes[1, 0].set_xlabel('Intensity')
    axes[1, 0].set_ylabel('Frequency')
    
    # 5. Segmentation classes
    unique_labels = np.unique(labels)
    label_counts = [np.sum(labels == label) for label in unique_labels]
    axes[1, 1].bar(unique_labels, label_counts)
    axes[1, 1].set_title('Segmentation Classes')
    axes[1, 1].set_xlabel('Class')
    axes[1, 1].set_ylabel('Pixel Count')
    
    # 6. Cancer by file
    cancer_by_file = df.groupby('file')['cancer'].max().reset_index()
    cancer_file_counts = cancer_by_file['cancer'].value_counts()
    file_labels = ['No Cancer Files', 'Cancer Files'] if len(cancer_file_counts) == 2 else ['Files']
    axes[1, 2].bar(range(len(cancer_file_counts)), cancer_file_counts.values)
    axes[1, 2].set_xticks(range(len(cancer_file_counts)))
    axes[1, 2].set_xticklabels(file_labels[:len(cancer_file_counts)])
    axes[1, 2].set_title('Files with/without Cancer')
    axes[1, 2].set_ylabel('Number of Files')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/eda_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()
    
    # Additional analysis
    print(f"\nDetailed Analysis:")
    severity_counts = df[df['cancer']==1].groupby('severity')['cancer'].count() if np.sum(df['cancer']==1) > 0 else pd.Series()
    print(f"Severity distribution in cancer cases: {dict(severity_counts)}")
    
    return df

# ============================================================================
# 3. MODEL ARCHITECTURE
# ============================================================================

def create_multi_output_unet(input_shape, num_seg_classes=3):
    """Create U-Net with segmentation + classification outputs"""
    
    inputs = layers.Input(input_shape)
    
    # Encoder
    enc1 = layers.Conv2D(64, 3, activation='relu', padding='same')(inputs)
    enc1 = layers.Conv2D(64, 3, activation='relu', padding='same')(enc1)
    pool1 = layers.MaxPooling2D(pool_size=(2, 2))(enc1)
    
    enc2 = layers.Conv2D(128, 3, activation='relu', padding='same')(pool1)
    enc2 = layers.Conv2D(128, 3, activation='relu', padding='same')(enc2)
    pool2 = layers.MaxPooling2D(pool_size=(2, 2))(enc2)
    
    enc3 = layers.Conv2D(256, 3, activation='relu', padding='same')(pool2)
    enc3 = layers.Conv2D(256, 3, activation='relu', padding='same')(enc3)
    pool3 = layers.MaxPooling2D(pool_size=(2, 2))(enc3)
    
    enc4 = layers.Conv2D(512, 3, activation='relu', padding='same')(pool3)
    enc4 = layers.Conv2D(512, 3, activation='relu', padding='same')(enc4)
    pool4 = layers.MaxPooling2D(pool_size=(2, 2))(enc4)
    
    # Bridge (bottleneck)
    bridge = layers.Conv2D(1024, 3, activation='relu', padding='same')(pool4)
    bridge = layers.Conv2D(1024, 3, activation='relu', padding='same')(bridge)
    
    # ========== SEGMENTATION DECODER ==========
    up4 = layers.UpSampling2D(size=(2, 2))(bridge)
    up4 = layers.concatenate([up4, enc4])
    dec4 = layers.Conv2D(512, 3, activation='relu', padding='same')(up4)
    dec4 = layers.Conv2D(512, 3, activation='relu', padding='same')(dec4)
    
    up3 = layers.UpSampling2D(size=(2, 2))(dec4)
    up3 = layers.concatenate([up3, enc3])
    dec3 = layers.Conv2D(256, 3, activation='relu', padding='same')(up3)
    dec3 = layers.Conv2D(256, 3, activation='relu', padding='same')(dec3)
    
    up2 = layers.UpSampling2D(size=(2, 2))(dec3)
    up2 = layers.concatenate([up2, enc2])
    dec2 = layers.Conv2D(128, 3, activation='relu', padding='same')(up2)
    dec2 = layers.Conv2D(128, 3, activation='relu', padding='same')(dec2)
    
    up1 = layers.UpSampling2D(size=(2, 2))(dec2)
    up1 = layers.concatenate([up1, enc1])
    dec1 = layers.Conv2D(64, 3, activation='relu', padding='same')(up1)
    dec1 = layers.Conv2D(64, 3, activation='relu', padding='same')(dec1)
    
    # Segmentation output
    segmentation_output = layers.Conv2D(num_seg_classes, 1, activation='softmax', name='segmentation')(dec1)
    
    # ========== CLASSIFICATION BRANCH ==========
    # Use bridge features + encoder features for classification
    gap_bridge = layers.GlobalAveragePooling2D()(bridge)
    gap_enc4 = layers.GlobalAveragePooling2D()(enc4)
    
    # Combine multi-scale features
    combined_features = layers.concatenate([gap_bridge, gap_enc4])
    
    # Classification layers
    dense1 = layers.Dense(512, activation='relu')(combined_features)
    dropout1 = layers.Dropout(0.5)(dense1)
    dense2 = layers.Dense(256, activation='relu')(dropout1)
    dropout2 = layers.Dropout(0.3)(dense2)
    dense3 = layers.Dense(128, activation='relu')(dropout2)
    dropout3 = layers.Dropout(0.2)(dense3)
    
    # Output layers
    cancer_output = layers.Dense(2, activation='softmax', name='cancer_detection')(dropout3)
    severity_output = layers.Dense(4, activation='softmax', name='severity_classification')(dropout3)
    
    model = models.Model(inputs=inputs, 
                        outputs=[segmentation_output, cancer_output, severity_output])
    
    return model

# ============================================================================
# 4. CUSTOM METRICS AND LOSS FUNCTIONS
# ============================================================================

def dice_coefficient(y_true, y_pred, smooth=1):
    """Dice coefficient for segmentation"""
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (tf.keras.backend.sum(y_true_f) + tf.keras.backend.sum(y_pred_f) + smooth)

def dice_loss(y_true, y_pred):
    """Dice loss for segmentation"""
    return 1 - dice_coefficient(y_true, y_pred)

def combined_segmentation_loss(y_true, y_pred):
    """Combined loss for segmentation"""
    return 0.5 * tf.keras.losses.categorical_crossentropy(y_true, y_pred) + 0.5 * dice_loss(y_true, y_pred)

# ============================================================================
# 5. TRAINING PIPELINE
# ============================================================================

def train_model(X_train, y_seg_train, y_cancer_train, y_severity_train,
                X_val, y_seg_val, y_cancer_val, y_severity_val):
    """Complete training pipeline"""
    
    print("="*60)
    print("TRAINING MULTI-OUTPUT U-NET MODEL")
    print("="*60)
    
    with strategy.scope():
        # Create model
        model = create_multi_output_unet(input_shape=(128, 128, 1), num_seg_classes=3)
        
        # Compile model
        model.compile(
            optimizer=Adam(learning_rate=1e-4),
            loss={
                'segmentation': combined_segmentation_loss,
                'cancer_detection': 'sparse_categorical_crossentropy',
                'severity_classification': 'sparse_categorical_crossentropy'
            },
            loss_weights={
                'segmentation': 1.0,
                'cancer_detection': 0.7,
                'severity_classification': 0.5
            },
            metrics={
                'segmentation': ['accuracy', MeanIoU(num_classes=3), dice_coefficient],
                'cancer_detection': ['accuracy'],
                'severity_classification': ['accuracy']
            }
        )
    
    # Print model summary
    model.summary()
    
    # Callbacks
    callbacks = [
        ModelCheckpoint(
            '/kaggle/working/best_model.h5',
            monitor='val_loss',
            save_best_only=True,
            save_weights_only=False,
            verbose=1
        ),
        EarlyStopping(
            monitor='val_loss',
            patience=15,
            restore_best_weights=True,
            verbose=1
        ),
        ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=8,
            min_lr=1e-7,
            verbose=1
        )
    ]
    
    # Calculate batch size
    batch_size = 16 * strategy.num_replicas_in_sync
    
    # Train model
    print(f"\nTraining with batch size: {batch_size}")
    history = model.fit(
        X_train,
        {
            'segmentation': y_seg_train,
            'cancer_detection': y_cancer_train,
            'severity_classification': y_severity_train
        },
        validation_data=(
            X_val,
            {
                'segmentation': y_seg_val,
                'cancer_detection': y_cancer_val,
                'severity_classification': y_severity_val
            }
        ),
        epochs=100,
        batch_size=batch_size,
        callbacks=callbacks,
        verbose=1
    )
    
    return model, history

# ============================================================================
# 6. EVALUATION AND METRICS
# ============================================================================

def evaluate_model(model, X_test, y_seg_test, y_cancer_test, y_severity_test):
    """Comprehensive model evaluation"""
    
    print("="*60)
    print("MODEL EVALUATION")
    print("="*60)
    
    # Get predictions
    predictions = model.predict(X_test, batch_size=32, verbose=1)
    seg_pred, cancer_pred, severity_pred = predictions
    
    # Convert predictions to classes
    cancer_pred_classes = np.argmax(cancer_pred, axis=1)
    severity_pred_classes = np.argmax(severity_pred, axis=1)
    seg_pred_classes = np.argmax(seg_pred, axis=-1)
    
    # FIX: Convert y_seg_test from one-hot to class indices if needed
    if len(y_seg_test.shape) == 4:  # One-hot encoded (N, H, W, C)
        y_seg_test_classes = np.argmax(y_seg_test, axis=-1)
    else:  # Already class indices (N, H, W)
        y_seg_test_classes = y_seg_test
    
    # Calculate metrics
    results = {}
    
    # 1. Cancer Detection Metrics
    print("\n1. CANCER DETECTION METRICS:")
    cancer_accuracy = np.mean(cancer_pred_classes == y_cancer_test)
    
    # Only calculate AUC if both classes are present
    if len(np.unique(y_cancer_test)) > 1:
        cancer_auc = roc_auc_score(y_cancer_test, cancer_pred[:, 1])
        print(f"AUC-ROC: {cancer_auc:.4f}")
    else:
        cancer_auc = 0.0
        print("AUC-ROC: N/A (only one class present)")
    
    print(f"Accuracy: {cancer_accuracy:.4f}")
    
    print("\nClassification Report:")
    print(classification_report(y_cancer_test, cancer_pred_classes, 
                              target_names=['No Cancer', 'Cancer']))
    
    # 2. Severity Classification Metrics
    print("\n2. SEVERITY CLASSIFICATION METRICS:")
    severity_accuracy = np.mean(severity_pred_classes == y_severity_test)
    print(f"Accuracy: {severity_accuracy:.4f}")
    
    # Get the actual classes present in the test data
    unique_classes = np.unique(y_severity_test)
    severity_names = ['None', 'Low', 'Medium', 'High']
    present_severity_names = [severity_names[i] for i in unique_classes]
    
    print("\nClassification Report:")
    print(classification_report(y_severity_test, severity_pred_classes, 
                              target_names=present_severity_names,
                              labels=unique_classes))
    
    # 3. Segmentation Metrics
    print("\n3. SEGMENTATION METRICS:")
    seg_accuracy = np.mean(seg_pred_classes == y_seg_test_classes)
    
    # Calculate IoU for each class
    seg_flat = y_seg_test_classes.flatten()
    seg_pred_flat = seg_pred_classes.flatten()
    
    iou_scores = []
    class_names = ['Background', 'Prostate', 'Lesion']
    for class_id in range(3):
        # Create binary masks for this class
        true_mask = (seg_flat == class_id)
        pred_mask = (seg_pred_flat == class_id)
        
        # Calculate IoU only if class exists in ground truth
        if np.sum(true_mask) > 0:
            iou = jaccard_score(true_mask, pred_mask)
            iou_scores.append(iou)
            print(f"IoU for {class_names[class_id]}: {iou:.4f}")
        else:
            iou_scores.append(0.0)
            print(f"IoU for {class_names[class_id]}: 0.0000 (class not present)")
    
    mean_iou = np.mean(iou_scores)
    
    print(f"\nPixel Accuracy: {seg_accuracy:.4f}")
    print(f"Mean IoU: {mean_iou:.4f}")
    
    # Calculate Dice coefficient for segmentation
    def calculate_dice(y_true, y_pred, class_id):
        y_true_class = (y_true == class_id).astype(int)
        y_pred_class = (y_pred == class_id).astype(int)
        intersection = np.sum(y_true_class * y_pred_class)
        dice = (2.0 * intersection) / (np.sum(y_true_class) + np.sum(y_pred_class) + 1e-8)
        return dice
    
    dice_scores = []
    for class_id in range(3):
        dice = calculate_dice(seg_flat, seg_pred_flat, class_id)
        dice_scores.append(dice)
        print(f"Dice for {class_names[class_id]}: {dice:.4f}")
    
    mean_dice = np.mean(dice_scores)
    print(f"Mean Dice: {mean_dice:.4f}")
    
    # Additional segmentation metrics
    precision, recall, f1, support = precision_recall_fscore_support(
        seg_flat, seg_pred_flat, labels=[0, 1, 2], average=None, zero_division=0
    )
    
    print(f"\nPer-class metrics:")
    for i in range(3):
        print(f"{class_names[i]}: Precision={precision[i]:.4f}, Recall={recall[i]:.4f}, F1={f1[i]:.4f}")
    
    # Store results
    results = {
        'cancer_accuracy': float(cancer_accuracy),
        'cancer_auc': float(cancer_auc),
        'severity_accuracy': float(severity_accuracy),
        'segmentation_accuracy': float(seg_accuracy),
        'mean_iou': float(mean_iou),
        'mean_dice': float(mean_dice),
        'iou_per_class': [float(x) for x in iou_scores],
        'dice_per_class': [float(x) for x in dice_scores],
        'precision_per_class': [float(x) for x in precision],
        'recall_per_class': [float(x) for x in recall],
        'f1_per_class': [float(x) for x in f1],
        'timestamp': datetime.now().isoformat()
    }
    
    return results, predictions

def plot_training_history(history):
    """Plot training curves"""
    
    fig, axes = plt.subplots(2, 3, figsize=(18, 10))
    fig.suptitle('Training History', fontsize=16)
    
    # Loss plots
    axes[0, 0].plot(history.history['loss'], label='Train')
    axes[0, 0].plot(history.history['val_loss'], label='Validation')
    axes[0, 0].set_title('Total Loss')
    axes[0, 0].legend()
    
    axes[0, 1].plot(history.history['cancer_detection_loss'], label='Train')
    axes[0, 1].plot(history.history['val_cancer_detection_loss'], label='Validation')
    axes[0, 1].set_title('Cancer Detection Loss')
    axes[0, 1].legend()
    
    axes[0, 2].plot(history.history['segmentation_loss'], label='Train')
    axes[0, 2].plot(history.history['val_segmentation_loss'], label='Validation')
    axes[0, 2].set_title('Segmentation Loss')
    axes[0, 2].legend()
    
    # Accuracy plots
    axes[1, 0].plot(history.history['cancer_detection_accuracy'], label='Train')
    axes[1, 0].plot(history.history['val_cancer_detection_accuracy'], label='Validation')
    axes[1, 0].set_title('Cancer Detection Accuracy')
    axes[1, 0].legend()
    
    axes[1, 1].plot(history.history['severity_classification_accuracy'], label='Train')
    axes[1, 1].plot(history.history['val_severity_classification_accuracy'], label='Validation')
    axes[1, 1].set_title('Severity Classification Accuracy')
    axes[1, 1].legend()
    
    axes[1, 2].plot(history.history['segmentation_accuracy'], label='Train')
    axes[1, 2].plot(history.history['val_segmentation_accuracy'], label='Validation')
    axes[1, 2].set_title('Segmentation Accuracy')
    axes[1, 2].legend()
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/training_history.png', dpi=300, bbox_inches='tight')
    plt.show()

def visualize_predictions(X_test, y_seg_test, y_cancer_test, y_severity_test, predictions, num_samples=6):
    """Visualize model predictions"""
    
    seg_pred, cancer_pred, severity_pred = predictions
    cancer_pred_classes = np.argmax(cancer_pred, axis=1)
    severity_pred_classes = np.argmax(severity_pred, axis=1)
    seg_pred_classes = np.argmax(seg_pred, axis=-1)
    
    # FIX: Convert y_seg_test from one-hot to class indices if needed
    if len(y_seg_test.shape) == 4:  # One-hot encoded (N, H, W, C)
        y_seg_test_classes = np.argmax(y_seg_test, axis=-1)
    else:  # Already class indices (N, H, W)
        y_seg_test_classes = y_seg_test
    
    fig, axes = plt.subplots(num_samples, 4, figsize=(16, 4*num_samples))
    if num_samples == 1:
        axes = axes.reshape(1, -1)
    fig.suptitle('Model Predictions', fontsize=16)
    
    severity_names = ['None', 'Low', 'Medium', 'High']
    cancer_names = ['No Cancer', 'Cancer']
    
    # Ensure num_samples doesn't exceed available test samples
    num_samples = min(num_samples, len(X_test))
    
    # Select diverse samples (some with cancer, some without if possible)
    cancer_indices = np.where(y_cancer_test == 1)[0]
    no_cancer_indices = np.where(y_cancer_test == 0)[0]
    
    selected_indices = []
    
    # Try to get a mix of cancer and no-cancer cases
    cancer_samples = min(num_samples // 2, len(cancer_indices))
    no_cancer_samples = min(num_samples - cancer_samples, len(no_cancer_indices))
    
    if len(cancer_indices) > 0:
        selected_indices.extend(np.random.choice(cancer_indices, cancer_samples, replace=False))
    if len(no_cancer_indices) > 0:
        selected_indices.extend(np.random.choice(no_cancer_indices, no_cancer_samples, replace=False))
    
    # If we still need more samples, add random ones
    while len(selected_indices) < num_samples:
        remaining_indices = list(set(range(len(X_test))) - set(selected_indices))
        if remaining_indices:
            selected_indices.append(np.random.choice(remaining_indices))
        else:
            break
    
    for i, idx in enumerate(selected_indices):
        # Original image
        axes[i, 0].imshow(X_test[idx].squeeze(), cmap='gray')
        axes[i, 0].set_title('Original Image')
        axes[i, 0].axis('off')
        
        # True segmentation
        axes[i, 1].imshow(y_seg_test_classes[idx], cmap='viridis', vmin=0, vmax=2)
        axes[i, 1].set_title('True Segmentation')
        axes[i, 1].axis('off')
        
        # Predicted segmentation
        axes[i, 2].imshow(seg_pred_classes[idx], cmap='viridis', vmin=0, vmax=2)
        axes[i, 2].set_title('Predicted Segmentation')
        axes[i, 2].axis('off')
        
        # Classification results
        true_cancer = cancer_names[y_cancer_test[idx]]
        pred_cancer = cancer_names[cancer_pred_classes[idx]]
        true_severity = severity_names[y_severity_test[idx]]
        pred_severity = severity_names[severity_pred_classes[idx]]
        
        # Add confidence scores
        cancer_confidence = cancer_pred[idx, cancer_pred_classes[idx]]
        severity_confidence = severity_pred[idx, severity_pred_classes[idx]]
        
        # Color code correct/incorrect predictions
        cancer_color = 'green' if y_cancer_test[idx] == cancer_pred_classes[idx] else 'red'
        severity_color = 'green' if y_severity_test[idx] == severity_pred_classes[idx] else 'red'
        
        axes[i, 3].text(0.05, 0.85, f"True Cancer: {true_cancer}", fontsize=10, 
                       transform=axes[i, 3].transAxes)
        axes[i, 3].text(0.05, 0.70, f"Pred Cancer: {pred_cancer} ({cancer_confidence:.3f})", 
                       fontsize=10, transform=axes[i, 3].transAxes, color=cancer_color)
        axes[i, 3].text(0.05, 0.55, f"True Severity: {true_severity}", fontsize=10, 
                       transform=axes[i, 3].transAxes)
        axes[i, 3].text(0.05, 0.40, f"Pred Severity: {pred_severity} ({severity_confidence:.3f})", 
                       fontsize=10, transform=axes[i, 3].transAxes, color=severity_color)
        axes[i, 3].set_title('Classification Results')
        axes[i, 3].axis('off')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/model_predictions.png', dpi=300, bbox_inches='tight')
    plt.show()

# ============================================================================
# 7. ADDITIONAL ANALYSIS FUNCTIONS
# ============================================================================

def plot_confusion_matrices(y_cancer_test, y_severity_test, predictions):
    """Plot confusion matrices for classification tasks"""
    
    seg_pred, cancer_pred, severity_pred = predictions
    cancer_pred_classes = np.argmax(cancer_pred, axis=1)
    severity_pred_classes = np.argmax(severity_pred, axis=1)
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))
    fig.suptitle('Confusion Matrices', fontsize=16)
    
    # Cancer detection confusion matrix
    cm_cancer = confusion_matrix(y_cancer_test, cancer_pred_classes)
    sns.heatmap(cm_cancer, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['No Cancer', 'Cancer'],
                yticklabels=['No Cancer', 'Cancer'],
                ax=axes[0])
    axes[0].set_title('Cancer Detection')
    axes[0].set_xlabel('Predicted')
    axes[0].set_ylabel('True')
    
    # Severity classification confusion matrix
    unique_severity = np.unique(y_severity_test)
    severity_names = ['None', 'Low', 'Medium', 'High']
    severity_labels = [severity_names[i] for i in unique_severity]
    
    cm_severity = confusion_matrix(y_severity_test, severity_pred_classes, labels=unique_severity)
    sns.heatmap(cm_severity, annot=True, fmt='d', cmap='Oranges',
                xticklabels=severity_labels,
                yticklabels=severity_labels,
                ax=axes[1])
    axes[1].set_title('Severity Classification')
    axes[1].set_xlabel('Predicted')
    axes[1].set_ylabel('True')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/confusion_matrices.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_roc_curves(y_cancer_test, predictions):
    """Plot ROC curves for cancer detection"""
    
    seg_pred, cancer_pred, severity_pred = predictions
    
    # Only plot if we have both classes
    if len(np.unique(y_cancer_test)) > 1:
        fpr, tpr, thresholds = roc_curve(y_cancer_test, cancer_pred[:, 1])
        auc_score = roc_auc_score(y_cancer_test, cancer_pred[:, 1])
        
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, linewidth=2, label=f'ROC Curve (AUC = {auc_score:.3f})')
        plt.plot([0, 1], [0, 1], 'k--', linewidth=1, label='Random')
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('ROC Curve for Cancer Detection')
        plt.legend()
        plt.grid(True, alpha=0.3)
        plt.savefig('/kaggle/working/roc_curve.png', dpi=300, bbox_inches='tight')
        plt.show()
    else:
        print("ROC curve not plotted: only one class present in test data")

def analyze_predictions_by_severity(y_cancer_test, y_severity_test, predictions):
    """Analyze model performance by severity levels"""
    
    seg_pred, cancer_pred, severity_pred = predictions
    cancer_pred_classes = np.argmax(cancer_pred, axis=1)
    severity_pred_classes = np.argmax(severity_pred, axis=1)
    
    # Create DataFrame for analysis
    analysis_df = pd.DataFrame({
        'true_cancer': y_cancer_test,
        'pred_cancer': cancer_pred_classes,
        'true_severity': y_severity_test,
        'pred_severity': severity_pred_classes,
        'cancer_confidence': cancer_pred[:, 1]
    })
    
    print("\n" + "="*50)
    print("ANALYSIS BY SEVERITY LEVEL")
    print("="*50)
    
    severity_names = ['None', 'Low', 'Medium', 'High']
    
    for severity in np.unique(y_severity_test):
        subset = analysis_df[analysis_df['true_severity'] == severity]
        if len(subset) > 0:
            cancer_acc = np.mean(subset['true_cancer'] == subset['pred_cancer'])
            severity_acc = np.mean(subset['true_severity'] == subset['pred_severity'])
            avg_confidence = np.mean(subset['cancer_confidence'])
            
            print(f"\nSeverity Level: {severity_names[severity]} (n={len(subset)})")
            print(f"  Cancer Detection Accuracy: {cancer_acc:.3f}")
            print(f"  Severity Classification Accuracy: {severity_acc:.3f}")
            print(f"  Average Cancer Confidence: {avg_confidence:.3f}")

# ============================================================================
# 8. MODEL INTERPRETATION FUNCTIONS
# ============================================================================

def visualize_feature_maps(model, X_test, layer_names=None, num_samples=3):
    """Visualize intermediate feature maps"""
    
    if layer_names is None:
        # Get some interesting layers
        layer_names = []
        for layer in model.layers:
            if 'conv2d' in layer.name and len(layer_names) < 6:
                layer_names.append(layer.name)
    
    # Create a model that outputs feature maps
    feature_extractor = Model(inputs=model.inputs, 
                             outputs=[model.get_layer(name).output for name in layer_names])
    
    # Get feature maps for a few samples
    sample_indices = np.random.choice(len(X_test), num_samples, replace=False)
    
    for idx in sample_indices:
        sample_input = X_test[idx:idx+1]
        feature_maps = feature_extractor.predict(sample_input, verbose=0)
        
        fig, axes = plt.subplots(2, len(layer_names), figsize=(4*len(layer_names), 8))
        if len(layer_names) == 1:
            axes = axes.reshape(-1, 1)
        fig.suptitle(f'Feature Maps for Sample {idx}', fontsize=16)
        
        # Show original image
        for j in range(len(layer_names)):
            axes[0, j].imshow(X_test[idx].squeeze(), cmap='gray')
            axes[0, j].set_title('Original')
            axes[0, j].axis('off')
        
        # Show feature maps
        for j, (layer_name, feature_map) in enumerate(zip(layer_names, feature_maps)):
            # Take the mean across all channels for visualization
            if len(feature_map.shape) == 4:
                mean_feature_map = np.mean(feature_map[0], axis=-1)
            else:
                mean_feature_map = feature_map[0]
            
            axes[1, j].imshow(mean_feature_map, cmap='viridis')
            axes[1, j].set_title(f'{layer_name}')
            axes[1, j].axis('off')
        
        plt.tight_layout()
        plt.savefig(f'/kaggle/working/feature_maps_sample_{idx}.png', dpi=300, bbox_inches='tight')
        plt.show()

# ============================================================================
# 9. MAIN EXECUTION PIPELINE
# ============================================================================

def main():
    """Main execution pipeline"""
    
    print("="*80)
    print("PROSTATE CANCER CLASSIFICATION AND SEGMENTATION PIPELINE")
    print("="*80)
    
    # 1. Load and preprocess data
    print("\n1. Loading and preprocessing data...")
    image_dir = '/kaggle/input/prostate-annotated-dataset-for-image-segmentation/Task05_Prostate/imagesTr'
    label_dir = '/kaggle/input/prostate-annotated-dataset-for-image-segmentation/Task05_Prostate/labelsTr'
    
    images, labels, cancer_labels, severity_labels, file_info = load_and_preprocess_data(image_dir, label_dir)
    
    # 2. EDA
    print("\n2. Performing EDA...")
    df = perform_eda(images, labels, cancer_labels, severity_labels, file_info)
    
    # 3. Prepare data for training
    print("\n3. Preparing data for training...")
    
    # Normalize images
    images = images / np.max(images)
    
    # Convert labels
    labels_categorical = tf.keras.utils.to_categorical(labels, num_classes=3)
    
    # Split data
    X_train, X_temp, y_seg_train, y_seg_temp, y_cancer_train, y_cancer_temp, y_severity_train, y_severity_temp = train_test_split(
        images, labels_categorical, cancer_labels, severity_labels, test_size=0.3, random_state=42, stratify=cancer_labels
    )
    
    X_val, X_test, y_seg_val, y_seg_test, y_cancer_val, y_cancer_test, y_severity_val, y_severity_test = train_test_split(
        X_temp, y_seg_temp, y_cancer_temp, y_severity_temp, test_size=0.5, random_state=42, stratify=y_cancer_temp
    )
    
    print(f"Training set: {len(X_train)} samples")
    print(f"Validation set: {len(X_val)} samples")
    print(f"Test set: {len(X_test)} samples")
    
    # 4. Train model
    print("\n4. Training model...")
    model, history = train_model(X_train, y_seg_train, y_cancer_train, y_severity_train,
                                X_val, y_seg_val, y_cancer_val, y_severity_val)
    
    # 5. Plot training history
    print("\n5. Plotting training history...")
    plot_training_history(history)
    
    # 6. Evaluate model
    print("\n6. Evaluating model...")
    results, predictions = evaluate_model(model, X_test, y_seg_test, y_cancer_test, y_severity_test)
    
    # 7. Additional analysis
    print("\n7. Performing additional analysis...")
    plot_confusion_matrices(y_cancer_test, y_severity_test, predictions)
    plot_roc_curves(y_cancer_test, predictions)
    analyze_predictions_by_severity(y_cancer_test, y_severity_test, predictions)
    
    # 8. Visualize predictions
    print("\n8. Visualizing predictions...")
    visualize_predictions(X_test, y_seg_test, y_cancer_test, y_severity_test, predictions)
    
    # 9. Feature map visualization (optional - can be memory intensive)
    print("\n9. Visualizing feature maps...")
    try:
        visualize_feature_maps(model, X_test, num_samples=2)
    except Exception as e:
        print(f"Feature map visualization skipped due to: {e}")
    
    # 10. Save everything
    print("\n10. Saving model and results...")
    
    # Save model
    model.save('/kaggle/working/prostate_multi_output_model.h5')
    
    # Save training history
    with open('/kaggle/working/training_history.json', 'w') as f:
        # Convert numpy arrays to lists for JSON serialization
        history_dict = {}
        for key, value in history.history.items():
            if isinstance(value, np.ndarray):
                history_dict[key] = value.tolist()
            elif isinstance(value, list):
                history_dict[key] = [float(v) if isinstance(v, (np.integer, np.floating)) else v for v in value]
            else:
                history_dict[key] = value
        json.dump(history_dict, f, indent=2)
    
    # Save evaluation results
    with open('/kaggle/working/evaluation_results.json', 'w') as f:
        json.dump(results, f, indent=2)
    
    # Save data splits for reproducibility
    np.savez('/kaggle/working/data_splits.npz',
             X_test=X_test, y_seg_test=y_seg_test, 
             y_cancer_test=y_cancer_test, y_severity_test=y_severity_test)
    
    # Save predictions for further analysis
    np.savez('/kaggle/working/predictions.npz',
             seg_pred=predictions[0], cancer_pred=predictions[1], severity_pred=predictions[2])
    
    print("\n" + "="*80)
    print("PIPELINE COMPLETED SUCCESSFULLY!")
    print("="*80)
    print(f"Model saved: /kaggle/working/prostate_multi_output_model.h5")
    print(f"Results saved: /kaggle/working/evaluation_results.json")
    print(f"History saved: /kaggle/working/training_history.json")
    print(f"Predictions saved: /kaggle/working/predictions.npz")
    print(f"Data splits saved: /kaggle/working/data_splits.npz")
    print(f"\nFinal Results:")
    print(f"  - Cancer Detection Accuracy: {results['cancer_accuracy']:.4f}")
    print(f"  - Cancer Detection AUC: {results['cancer_auc']:.4f}")
    print(f"  - Severity Classification Accuracy: {results['severity_accuracy']:.4f}")
    print(f"  - Segmentation Accuracy: {results['segmentation_accuracy']:.4f}")
    print(f"  - Mean IoU: {results['mean_iou']:.4f}")
    print(f"  - Mean Dice: {results['mean_dice']:.4f}")

# ============================================================================
# 10. UTILITY FUNCTIONS FOR POST-TRAINING ANALYSIS
# ============================================================================

def load_saved_model_and_analyze(model_path='/kaggle/working/prostate_multi_output_model.h5',
                                data_path='/kaggle/working/data_splits.npz',
                                predictions_path='/kaggle/working/predictions.npz'):
    """Load saved model and perform additional analysis"""
    
    print("Loading saved model and data...")
    
    # Load model
    model = tf.keras.models.load_model(model_path, 
                                      custom_objects={
                                          'dice_coefficient': dice_coefficient,
                                          'combined_segmentation_loss': combined_segmentation_loss
                                      })
    
    # Load data
    data = np.load(data_path)
    X_test = data['X_test']
    y_seg_test = data['y_seg_test']
    y_cancer_test = data['y_cancer_test']
    y_severity_test = data['y_severity_test']
    
    # Load predictions
    pred_data = np.load(predictions_path)
    predictions = (pred_data['seg_pred'], pred_data['cancer_pred'], pred_data['severity_pred'])
    
    print("Performing additional analysis...")
    
    # Re-run evaluation
    results, _ = evaluate_model(model, X_test, y_seg_test, y_cancer_test, y_severity_test)
    
    # Additional visualizations
    plot_confusion_matrices(y_cancer_test, y_severity_test, predictions)
    plot_roc_curves(y_cancer_test, predictions)
    analyze_predictions_by_severity(y_cancer_test, y_severity_test, predictions)
    visualize_predictions(X_test, y_seg_test, y_cancer_test, y_severity_test, predictions)
    
    return model, results, predictions

if __name__ == "__main__":
    main()
```
